### Imports

We begin by importing some relevant libraries. We import `ortools` to formulate and solve the LP, and we import `time` to time the entire process of supplying the constraints solving the LP.

```python
from ortools.linear_solver import pywraplp as OR
import time
```

```python
start_time = time.time()

m = OR.Solver('LP Solver', OR.Solver.GLOP_LINEAR_PROGRAMMING)

# add variable x
x = {}
N = 2
for i in range(1, N+1):
    x[i] = m.NumVar(0, m.infinity(), f'x[{i}]')

# add constraints
m.Add(4*x[1] + 3*x[2] <= 12)
m.Add(2*x[1] + 4*x[2] <= 12)
    
# set objective
objective = x[1] + x[2]
m.Maximize(objective)

status = m.Solve()

end_time = time.time()

diff = time.gmtime(end_time - start_time)
print('\n[Total time used: {} minutes, {} seconds]'.format(diff.tm_min, diff.tm_sec))

print('Objective:', m.Objective().Value())

if status == OR.Solver.OPTIMAL:
    print("Solution found and is optimal.")
    # Access solution values: solver.Objective().Value(), x.solution_value()
elif status == OR.Solver.FEASIBLE:
    print("Solution found, but may not be optimal.")
elif status == OR.Solver.INFEASIBLE:
    print("The problem does not have a feasible solution.")
elif status == OR.Solver.UNBOUNDED:
    print("The problem is unbounded.")
else:
    print("Solver status: Unknown or Error")
```

```python
# optimal solution
for i in x:
    print(f'{x[i]} = {x[i].solution_value()}')
```
